## Perhitungan Statistik


---


1. Rata-Rata
2. Standar Deviasi
3. Skewness
4. Median
5. Energi
6. Min
7. Max

```python
# Fungsi untuk menghitung energi
def calculate_energy(column):
    return np.sum(np.square(column))

# Inisialisasi dictionary untuk menyimpan hasil perhitungan
statistics_dict = {}

# Perhitungan statistik untuk setiap kolom
for column in df.columns:
    statistics_dict[column] = [
        df[column].mean(),
        df[column].std(),
        skew(df[column]),
        df[column].median(),
        calculate_energy(df[column]),
        df[column].min(),
        df[column].max()
    ]

# Ubah dictionary menjadi DataFrame
statistics_df = pd.DataFrame.from_dict(statistics_dict, orient='index', columns=['Rata-rata', 'Standar Deviasi', 'Skewness', 'Median', 'Energi', 'Min', 'Max'])

# Tampilkan dataframe hasil perhitungan
print("Statistik untuk setiap kolom:")
print(statistics_df)
# Konversi dataframe menjadi format tabel
table = tabulate(statistics_df, headers='keys', tablefmt='pretty')

# Menampilkan tabel
print("Statistik untuk setiap kolom:")
print(table)
```
