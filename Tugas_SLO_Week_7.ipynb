## Perhitungan Statistik


---


1. Rata-Rata
2. Standar Deviasi
3. Skewness
4. Median
5. Energi
6. Min
7. Max

```python
# Menghitung statistik
mean = df.mean()
std = df.std()
skewness = df.skew()
median = df.median()
energy = (df ** 2).sum()
min_val = df.min()
max_val = df.max()

# Menyusun statistik menjadi DataFrame
statistics_df = pd.DataFrame({
    'Rata-Rata': mean,
    'Standar Deviasi': std,
    'Skewness': skewness,
    'Median': median,
    'Energi': energy,
    'Min': min_val,
    'Max': max_val
})

# Menampilkan statistik DataFrame
statistics_df
```

## Plot Sinyal Plot dengan multiplot 2x3


---

Sinyal yang dipilih yaitu data 953, 915, 989, 957, 942, 945



```python
# Plot sinyal dari data0.csv
plt.figure(figsize=(12, 8))
for i, column in enumerate(df.columns[:6]):
    plt.subplot(2, 3, i+1)
    plt.plot(df[column])
    plt.title(column)
plt.tight_layout()
plt.show()
```

## Plot Scatterplot dari hasil no. 3 untuk Rata-rata dan Standar Deviasi

```python
plt.figure(figsize=(10, 6))

for i in range(0, len(mean), 20):
    plt.scatter(mean[i:i+20], std[i:i+20], label=f'Data {i+1}-{i+20}')

plt.xlabel('Mean')
plt.ylabel('Standar Deviasi')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title('Plot Scatter untuk Rata-Rata dan Standar Deviasi')

plt.show()
```

## Plot Scatterplot untuk Skewness dan Median

```python
plt.figure(figsize=(10, 6))

for i in range(0, len(skewness), 20):
    plt.scatter(skewness[i:i+20], median[i:i+20], label=f'Data {i+1}-{i+20}')

plt.xlabel('Skewness')
plt.ylabel('Median')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title('Plot Scatter untuk Skewness dan Median')

plt.show()
```
